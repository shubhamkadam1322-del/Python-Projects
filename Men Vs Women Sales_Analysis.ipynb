# Exploratory Data Analysis

### Gender

```python
ax = sns.countplot(x= "Gender", hue="Gender", data=df, width=0.4, palette = ['#1f77b4', '#d62728'], legend=False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender', hue="Gender", y= 'Amount' ,data = sales_gen, width=0.4, palette = ['#008080', '#FF7F0E'], legend=False)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', width= 0.7, legend=False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group', hue="Age Group", y= 'Amount' ,data = sales_age, width=0.4, palette = ['#001f3f', '#800080', "#001f3f",  '#004c4c', '#4B0082', '#d62728', '#2F4F4F'], legend=False)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State', hue="State", y= 'Orders', palette = ['#1f77b4', '#264653', '#3b0a45', '#d62728', '#800080',  '#004c4c', '#4B0082', '#d62728', '#2F4F4F', '#001f3f'], legend=False)
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State', hue= "State", y= 'Amount', palette = ['#1f77b4', '#264653', '#3b0a45', '#d62728', '#800080',  '#004c4c', '#4B0082', '#d62728', '#2F4F4F', '#001f3f'], legend= False)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status', hue= "Marital_Status", width= 0.3, palette = ['#1f77b4', "#3b0a45"], legend=False)

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status', width=0.5,  y= 'Amount', hue='Gender', palette = ['#1f77b4', "#3b0a45"], legend=True)
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data = df, x = 'Occupation', hue="Occupation", palette = ['#1f77b4', '#264653', '#3b0a45', '#d62728', '#800080',  '#004c4c', '#4B0082', '#d62728', '#2F4F4F', '#001f3f', '#bb3e03', '#14213d', '#2a3d66', '#673147','#8338ec'], legend=False)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation', hue="Occupation", y= 'Amount', palette = ['#1f77b4', '#264653', '#3b0a45', '#d62728', '#800080',  '#004c4c', '#4B0082', '#d62728', '#2F4F4F', '#001f3f', '#bb3e03', '#14213d', '#2a3d66', '#673147','#8338ec'], legend=False)
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category', hue="Product_Category", palette = ['#1f77b4', '#264653', '#3b0a45', '#d62728', '#800080',  '#004c4c', '#4B0082', '#d62728', '#2F4F4F', '#001f3f', '#bb3e03', '#14213d', '#2a3d66', '#673147','#8338ec','#cb997e','#7b2cbf','#4B0082'], legend=False)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category', hue = "Product_Category",y= 'Amount', palette = ['#1f77b4', '#264653', '#3b0a45', '#d62728', '#800080',  '#004c4c', '#4B0082', '#720026', '#2F4F4F', '#001f3f'], legend=False)
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID', hue="Product_ID",y= 'Orders', palette = ['#1f77b4', '#264653', '#3b0a45', '#d62728', '#800080',  '#004c4c', '#4B0082', '#720026', '#2F4F4F', '#001f3f'], legend=False)
```

```python
# top 10 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))

df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(
    kind='bar',
    color=['#1f77b4', '#264653', '#3b0a45', '#d62728', '#800080', '#004c4c', '#4B0082', '#720026', '#2F4F4F', '#001f3f'],
    legend=False,
    ax=ax1
)
```
